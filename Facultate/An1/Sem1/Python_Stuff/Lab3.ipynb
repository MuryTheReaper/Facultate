```python
#LAB 3
from math import sqrt
def ui():

    #printeaza interfata

    print("1: Alege sirul pe care lucram (numere intregi)")
    print("2: Gaseste secventele de lungime maxima care contin doar numere prime")
    print("3: Gaseste secventele de lungime maxima in care oricare doua elemente consecutive difera printr-un numar prim")
    print("4: Gaseste secventele de lungime maxima in care toate elementele sunt egale")
    print("0: Inchide programul")
    print()
    print()

def convSTRtoLST(sir:str):
    
    #transforma un string intr-o lista cu separator caracterul spatiu
    #primeste un string
    #returneaza o lista

    sir=sir.split()
    sir=list(map(int,sir))
    return sir

def prim(n:int):

    #verifica daca un numar este prim
    #primeste un nr intreg
    #returneaza True daca e prim, Flase daca nu
    if(n==2):
            return True
    else:
        if(n<2 or n%2==0):
            return False
        else:
            for i in range(3, int(sqrt(n))+1, 2):
                if(n%i==0):
                    return False
    return True

def secvmax_nr_prim(sir:list):

    #secvente maximale foarmate doar din numere prime
    #primeste o lista
    #returneaza o matrice care are secventele de sir cerute sau False daca nu exista
    
    lmax=0
    l=0
    rez=[[]]
    del rez[0]
    for el in sir:
        if prim(el)==True:
            l+=1
        else:
            if l>lmax:
                lmax=l
            l=0
    if l>lmax:
        lmax=l

    l=0
    if lmax>0:
        for i,el in enumerate(sir):
            if prim(el)==True:
                l+=1
            else:
                if l==lmax:
                    rez.append(sir[i-lmax:i])
                l=0
        if l==lmax:
             rez.append(sir[i-lmax+1:i+1])
        return rez
    else:
        return False

def secvmax_dif_de_nr_nrprim(sir:list):
    
    #secvente maximale cu diferenta dintre ele sa fie numar prim
    #primeste o lista
    #returneaza o matrice care are secventele de sir cerute sau False daca nu exista
    
    lmax=1
    l=1
    last=sir[0]
    rez=[[]]
    del rez[0]
    for el in sir:
        if prim(abs(last-el))==True:
            l+=1
        else:
            if l>lmax:
                lmax=l
            l=1
        last=el

    if l>lmax:
        lmax=l

    l=1
    if lmax>1:
        last=sir[0]
        for i,el in enumerate(sir):
            if prim(abs(last-el))==True:
                l+=1
            else:
                if l==lmax:
                    rez.append(sir[i-lmax:i])
                l=1
            last=el
        if l==lmax:
             rez.append(sir[i-lmax+1:i+1])
        return rez
    else:
        return False

def secvmax_nr_egale(sir:list):
    lmax=1
    l=1
    last=sir[0]
    rez=[[]]
    del rez[0]
    for el in sir[1:]:
        if last==el:
            l+=1
        else:
            if l>lmax:
                lmax=l
            l=1
        last=el
    if l>lmax:
        lmax=l
    l=1
    if lmax>1:
        last=sir[0]
        for i,el in enumerate(sir[1:]):
            if last==el:
                l+=1
            else:
                if l==lmax:
                    rez.append(sir[i-lmax+1:i+1])
                l=1
            last=el
        if l==lmax:
            rez.append(sir[i-lmax+2:i+2])
        return rez
    else:
        return False

def afis(sir,operatie):

    #afisare standar pentru o operatie data

    if not sir:
        print("Nu avem sir de prelucrat")
    else:
        if operatie==2:
            rez=secvmax_nr_prim(sir)
        else:
            if operatie==3:
                rez=secvmax_dif_de_nr_nrprim(sir)
            elif operatie==4:
                rez=secvmax_nr_egale(sir)
        if rez==False:
            print("Nu sunt secvente care satisfac cerinta")
        else:
            print("Secventele sunt: ", rez)
            #print(rez)
    print()

def test():

    #teste pentru prim
    assert prim(6)==False
    assert prim(5)==True
    assert prim(2)==True
    assert prim(abs(4-2))==True
    assert prim(3)==True

    #teste pentru operatia 1
    assert secvmax_nr_prim([2,3,5,9,0,11,8,5,3,2])==[[2,3,5],[5,3,2]]
    assert secvmax_nr_prim([4,6,8,10])==False
    
    #teste pentru operatia 2
    assert secvmax_dif_de_nr_nrprim([4,6,8,9,9,9,10,12,14])==[[4,6,8],[10,12,14]]
    assert secvmax_dif_de_nr_nrprim([11,8,13])==[[11,8,13]]

    #teste pentru operatia 3
    assert secvmax_nr_egale([2,3,3,3,5,4,4,4])==[[3,3,3,],[4,4,4]]
    assert secvmax_nr_egale([0,-1,-1,0])==[[-1,-1]]
    assert secvmax_nr_egale([0,-1,-1,-1])==[[-1,-1,-1]]


def main():
  
    #functia care porneste programul
    while True:
        ui()
        n=int(input("Ce operatie facem? "))
        if n==0:
            break
        else:
            if n==1:

                sir=(input("Introduceti sirul (format: 1 2 3 4): "))
                sir=convSTRtoLST(sir)
                print("Acesta este sirul:", sir)

            else:
                if n==2:

                    afis(sir,n)

                else:
                    if n==3:

                        afis(sir,n)
                    else:
                        if n==4:

                            afis(sir,n)

test()
main()
```
