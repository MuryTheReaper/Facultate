```python
#1a
reset()
t,w0=var('t,w0')
x=function('x')(t)
equ=diff(x,t,2)+w0^2*x==0
assume(w0>0)
show(desolve(equ,[x,t]))
```

```python
#1b
x0,v0=var('x0,v0')
sol(t,w0,x0,v0)=desolve(equ,[x,t],[0,x0,v0])
show(sol(t,w0,x0,v0))
```

```python
#1c
w0=sqrt(9.81/0.3924)
x0=0.15
v0=0
plot(sol(t,w0,x0,v0),t,0,3)
```

```python
#2a
reset()
t,w0,Y=var('t,w0,Y')
x=function('x')(t)
equ=diff(x,t,2)+Y*diff(x,t)+w0^2*x==0
assume(Y^2>4*w0^2)
#show(equ)
show(desolve(equ,[x,t]))
```

```python
#3a
reset()
t,w0,F0,w=var('t,w0,F0,w')
x=function('x')(t)
equ(w0,F0,w)=diff(x,t,2)+w0^2*x==F0*cos(w*t)
assume(w0>0)
assume(w0!=w)
show(desolve(equ,[x,t]))
```

```python
#3b
reset()
t,w0,F0,w=var('t,w0,F0,w')
x=function('x')(t)
equ(w0,F0,w)=diff(x,t,2)+w0^2*x==F0*cos(w*t)
sol(t)=desolve(equ(5,2,5.5),[x,t],[0,0,0])
plot(sol(t),t,0,5)
```

```python
#3d
reset()
t,w0,F0,w=var('t,w0,F0,w')
x=function('x')(t)
equ(w0,F0,w)=diff(x,t,2)+w0^2*x==F0*cos(w*t)
sol(t)=desolve(equ(5,2,5),[x,t],[0,0,0])
plot(sol(t),t,0,5)
```

```python
#3e
reset()
t,w0,F0,w=var('t,w0,F0,w')
x=function('x')(t)
equ(w0,F0,w)=diff(x,t,2)+w0^2*x==F0*cos(w*t)
assume(w0>0)
sol(t,w)=desolve(equ(w0,F0,w),[x,t],[0,0,0])
show(limit(sol(t,w),w=w0))
```
