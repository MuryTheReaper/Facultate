```python
#1a
reset()
t=var('t')
x=function('x')(t)

s=var('s')
f=function('f')(s)
f(s)=s^2-2*s

eqp=solve(f(s)==0,s)
eqp
```

```python
#1a, aproximatie
x1=eqp[0].rhs()
x2=eqp[1].rhs()
```

```python
diff(f,s)(x1) #local asimptotic stabil
```

```python
diff(f,s)(x2) #instabil
```

```python
#1a, grafic
sf=plot_slope_field(f(s),(t,-5,5),(s,-3,3),headaxislength=3, headlength=4, color='red')
sf
```

```python
deq=diff(x,t)==f(x)
g1=desolve_rk4(deq,x,[0,-1],step=0.1, end_points=[-5,5],output='plot', color='black')
g2=desolve_rk4(deq,x,[0,0],step=0.1, end_points=[-5,5],output='plot', color='blue')
g3=desolve_rk4(deq,x,[0,1],step=0.1, end_points=[-5,5],output='plot', color='black')
g4=desolve_rk4(deq,x,[0,2],step=0.1, end_points=[-5,5],output='plot', color='blue')
g5=desolve_rk4(deq,x,[0,3],step=0.1, end_points=[-5,5],output='plot', color='black')
g=sf+g1+g2+g3+g4+g5
g.show(ymin=-3,ymax=3)
```
