```python
reset()
x,k=var('x,k')
y=function('y')(x)
equ=diff(y,x)+k/x*y==x^3
sol(x,k)=desolve(equ,[y,x])
sol(x,k)
```

```python
_C=var('_C')
sol1(x,k)=sol(x,k).substitute(_C==1)
plot([sol1(x,1),sol1(x,2),sol1(x,3)],x,-3,3,detect_poles='True',ymin=-50,ymax=50)
```
