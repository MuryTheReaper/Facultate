```python
#1a
reset()
t,k,x0=var('t,k,x0')
x=function('x')(t)
equ=diff(x,t)==-k*x
sol(t,k,x0)=desolve(equ,[x,t],[0,x0])
show(sol)
```

```python
#1b
plot([sol(t,0.1,10),sol(t,0.2,10),sol(t,0.3,10)],t,0,3)
```

```python
#1c
t12=var('t12')
eq=sol(t12,k,x0)==x0/2
kC14(t12)=solve(eq,k)
show(kC14(5730))
```

```python
#1d
kC14=log(2)/5730
T=var('T')
ec2=sol(T,kC14,1)==0.2
solve(ec2,T)
```

```python
#2a
reset()
t,k,Tm,T0=var('t,k,Tm,T0')
T=function('T')(t)
equ=diff(T,t)==-k*(T-Tm)
sol(t,k,T0,Tm)=desolve(equ,[T,t],[0,T0])
show(sol)
```

```python
plot([sol(t,0.3,25,15),sol(t,0.3,5,15),sol(t,0.3,15,15)],t,0,3)
```
