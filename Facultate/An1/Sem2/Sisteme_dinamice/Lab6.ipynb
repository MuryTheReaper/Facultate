```python
#1b
reset()
reset()
t=var('t')
x=function('x')(t)

s=var('s')
f=function('f')(s)

f(s)=s*(s-1)*(s-2)

eqp=solve(f(s)==0,s,to_poly_solve='force')
eqp
```
